```python
import sys
import nltk
```

```python
def map_task(text_):
    w_counts = {}
    for line in text_:
        words = line.split()
        for word in words:
            w_counts[word] = w_counts.get(word, 0) + 1
    print w_counts
```

```python
a = nltk.corpus.gutenberg.words('austen-sense.txt')
```

```python
mapped = map_task(a).sort()
```

```python
def wcount(prev_key, counts):
    if prev_key is not None:
        print prev_key + " count is " + str(counts)
        
```

```python
def reduce_task(text_):
    prev_key = None
    counts = 0
    for line in text_:
        key, value = line.split("\t",1)
        if key!=prev_key:
            wcount(prev_key, counts)
            prev_key = key
            counts = 0
        counts+= eval (value)
    wcount(prev_key, counts)
```

```python
reduce_task(mapped)
```
